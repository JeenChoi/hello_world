```
na_values = ['NO CLUE', 'N/A', '0']
requests = pd.read_csv('https://data.cityofnewyork.us/resource/fhrw-4uyv.csv',
                      na_values=na_values,
                      dtype={'city': str})
```

```
def fix_city_name(city):
  # Truncate everything to length 10
  cities2 = cities.str.slice(0,10)
  
  #Set Jamaica to nan
  Jamaica_city = cities == 'Jamaica'
  cities2[Jamaica_city] = np.nan
  
  return cities
```

```
requests['city'] = fix_city_name(requests['city'])
```
